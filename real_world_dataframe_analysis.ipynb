No NaN or any invalid values in our data also. Since our data is ok and doesn't need any fix in values , we can proceed in our analysis. We will be adding a column to our data that will take account the combination of the economy in a city with its taxation and cost of living,business freedom and venture capital to get a better idea of the struggle(or not) that it takes to live there.

```python
#first we initialize our weights , those will help us with our calculation of the values for our new column
# I normalize each column by the mean of its value in our data,that is our weighted value
cost_of_living_mean=cities_df['Cost of Living'].mean()
cities_df['weighted_cost']=cities_df['Cost of Living']/cost_of_living_mean

taxation_mean=cities_df['Taxation'].mean()
cities_df['weighted_taxation']=cities_df['Taxation']/taxation_mean

economy_mean=cities_df['Economy'].mean()
cities_df['weighted_economy']=cities_df['Economy']/economy_mean

business_freedom_mean=cities_df['Business Freedom'].mean()
cities_df['weighted_business_freedom']=cities_df['Business Freedom']/business_freedom_mean

venture_capital_mean=cities_df['Venture Capital'].mean()
cities_df['weighted_venture_capital']=cities_df['Venture Capital']/venture_capital_mean


cities_df['Financial_Viability']=cities_df['weighted_economy']*cities_df['Economy']+\
cities_df['weighted_cost']*cities_df['Cost of Living']+cities_df['weighted_taxation']*cities_df['Taxation']+\
cities_df['Business Freedom']*cities_df['weighted_business_freedom']+\
cities_df['Venture Capital']*cities_df['weighted_venture_capital']


cities_df['Financial_Viability']=10*(cities_df['Financial_Viability']/cities_df['Financial_Viability'].max())
cities_df.sort_values('Financial_Viability',ascending=False).head(10)
cities_df.drop(columns=['weighted_cost'], inplace=True)
cities_df.drop(columns=['weighted_economy'], inplace=True)
cities_df.drop(columns=['weighted_taxation'], inplace=True)
cities_df.drop(columns=['weighted_business_freedom'], inplace=True)
cities_df.drop(columns=['weighted_venture_capital'], inplace=True)
```

#### now the calculation follows and  finally I divide by the maximum value and multiply by 10 to bring it to maximum 10 value like all the other columns.

```python
#These are the top 10 cities to live according to my new column 'Financial_Viability', the first 4 ones scored a 10 
#due to the fact that they had the maximum value for 'Financial_Viability' and since we divided by the maximum value 
#they had to have the perfect ten
cities_df.sort_values('Financial_Viability',ascending=False).head(10)
```

#### Q1: Which cities score moderate when it come to financial viability, meaning they score close to the total mean value in that criteria?

#### We first append to a list all the cities with close to mean values(range 0.1 from it) and then search for those cities to append theon another list , so that we concatenate all of them into one for our final result

```python
cities_df.shape#=(266,22)

mean=cities_df.Financial_Viability.mean()#=4.1659
result1=list()
for cities in cities_df.Financial_Viability:
    if (cities<mean+0.1 and cities>mean-0.1):
            result1.append(cities)

result=[]
for i in result1:
    result.append(cities_df[cities_df['Financial_Viability']==i])

pd.concat(result,ignore_index=True)
```

### Almost astonishing! 20 cities in the range of 0,1 from the mean value of the financial viability . Just shows that we do not have a big standard deviation in this distribution.
